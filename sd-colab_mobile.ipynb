```python
#@markdown # Checkpoint 
#@markdown  *Install Model / Checkpoint.*
!cd /content/AUTOMATIC1111-mobile/

print(f"{H}Menginstall model checkpoint...{D}")
sleep(8)

#-model name
Model = "Default" #@param ["Default", "abyssorange_mix3", "anypastel"]
#-model checkpoint
model_urls1 = {
    "abyssorange_mix3": 'https://huggingface.co/WarriorMama777/OrangeMixs/resolve/main/Models/AbyssOrangeMix3/AOM3_orangemixs.safetensors',
    "anypastel": 'https://civitai.com/api/download/models/17701',
}

#-VAE
model_urls2 = {
    "abyssorange_mix3": 'https://huggingface.co/WarriorMama777/OrangeMixs/resolve/main/VAEs/orangemix.vae.pt',
    "anypastel": 'https://huggingface.co/nolanaatama/nypstl/resolve/main/nypstl.vae.pt'
}

#-checkpoint dir
output_path1 = "/content/AUTOMATIC1111-mobile/models/Stable-diffusion/"
#-vae dir
output_path2 = "/content/AUTOMATIC1111-mobile/models/VAE/"

os.makedirs(output_path1, exist_ok=True)  
os.makedirs(output_path2, exist_ok=True)  

def download_model(model_url1, model_url2, output_path1, output_path2):
    filename = os.path.basename(model_url1)
    filename2 = os.path.basename(model_url2)
    output_file = os.path.join(output_path1, filename)
    output_file2 = os.path.join(output_path2, filename2)
    urllib.request.urlretrieve(model_url1, output_file)
    urllib.request.urlretrieve(model_url2, output_file2)
    return filename, filename2
    
if Model != "Default":
    model_url1 = model_urls1.get(Model)
    model_url2 = model_urls2.get(Model)
    download_model(model_url1, model_url2, output_path1, output_path2)
    print(f"{D}Checkpoint {B}{Model} {D}berhasil terpasang!...{B}\nLanjutkan proses berikutnya.{D}")
else:
    #clear_output()
    print(f"{D}\nAnda memilih {M}Default{D}\nTidak ada model terinstall...\n{H}majicmix_realistic5.safetensors{D} dipilih sebagai default.\n", f"{B}\nLanjutkan proses berikutnya,{D}\nAtau anda dapat memilih menginstall model kembali sebelum melanjutkan cell berikutnya.{D}")
```
